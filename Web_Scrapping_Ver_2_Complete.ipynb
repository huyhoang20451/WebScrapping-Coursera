```python
import pandas as pd

# URL of the webpage containing the Tesla Revenue table
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Use pandas read_html to directly extract tables from the webpage
tables = pd.read_html(url)

# Assuming the desired table is the first one, you can select it
tesla_revenue = tables[1]

# Rename columns for clarity
tesla_revenue.columns = ["Date", "Revenue (Millions of US $)"]

# Remove the rows with NaN values in the 'Revenue' column
tesla_revenue = tesla_revenue.dropna(subset=["Revenue (Millions of US $)"])

# Clean up the 'Revenue' column (remove commas and dollar signs)
tesla_revenue["Revenue (Millions of US $)"] = tesla_revenue["Revenue (Millions of US $)"].str.replace(',|\$', '', regex=True)

# Display the cleaned DataFrame
print(tesla_revenue)
```

```python
# Convert column names to lowercase and strip extra spaces
tesla_revenue.columns = tesla_revenue.columns.str.strip().str.lower()

# Clean up the 'revenue' column (remove commas and dollar signs)
tesla_revenue["revenue (millions of us $)"] = tesla_revenue["revenue (millions of us $)"].str.replace(',|\$', '', regex=True)

# Display the cleaned DataFrame
print(tesla_revenue)
```

```python
import pandas as pd

# URL of the webpage containing the Tesla Revenue table
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Use pandas read_html to directly extract tables from the webpage
tables = pd.read_html(url)

# Assuming the desired table is the first one, you can select it
gme_revenue = tables[1]

# Rename columns for clarity
gme_revenue.columns = ["Date", "GameStop Revenue"]

# Remove the rows with NaN values in the 'Revenue' column
gme_revenue = gme_revenue.dropna(subset=["GameStop Revenue"])

# Clean up the 'Revenue' column (remove commas and dollar signs)
gme_revenue["GameStop Revenue"] = gme_revenue["GameStop Revenue"].str.replace(',|\$', '', regex=True)

# Display the cleaned DataFrame
print(gme_revenue)
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def make_graph(stock_data_filename, revenue_data_filename, stock):
    # Load CSV files into DataFrames
    stock_data = pd.read_csv(stock_data_filename)
    revenue_data = pd.read_csv(revenue_data_filename)

    # Convert 'Date' columns to datetime objects
    stock_data['Date'] = pd.to_datetime(stock_data['date'], infer_datetime_format=True)
    revenue_data['Date'] = pd.to_datetime(revenue_data['Date'], infer_datetime_format=True)

    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=0.3)

    stock_data_specific = stock_data[stock_data['date'] <= '2021-04-16']
    revenue_data_specific = revenue_data[revenue_data['Date'] <= '2021-04-30']

    fig.add_trace(go.Scatter(x=stock_data_specific['Date'], y=stock_data_specific['revenue (millions of us $)'].astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=revenue_data_specific['Date'], y=revenue_data_specific['GameStop Revenue'].astype("float"), name="Revenue"), row=2, col=1)

    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)

    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    fig.show()

# Example usage:
make_graph('tesla_revenue.csv', 'gme_revenue.csv', 'Tesla')
```

```python
make_graph('tesla_revenue.csv', 'gme_revenue.csv', 'GameStop')
```
